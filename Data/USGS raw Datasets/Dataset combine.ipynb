# Combine various datasets


```python
import pandas as pd

# Read all the dataframes
dataframes = [
    pd.read_csv('01 renewable-share-energy.csv'),
    pd.read_csv('02 modern-renewable-energy-consumption.csv'),
    pd.read_csv('03 modern-renewable-prod.csv'),
    pd.read_csv('04 share-electricity-renewables.csv'),
    pd.read_csv('05 hydropower-consumption.csv'),
    pd.read_csv('06 hydro-share-energy.csv'),
    pd.read_csv('07 share-electricity-hydro.csv'),
    pd.read_csv('08 wind-generation.csv'),
    pd.read_csv('09 cumulative-installed-wind-energy-capacity-gigawatts.csv'),
    pd.read_csv('10 wind-share-energy.csv'),
    pd.read_csv('11 share-electricity-wind.csv'),
    pd.read_csv('12 solar-energy-consumption.csv'),
    pd.read_csv('13 installed-solar-PV-capacity.csv'),
    pd.read_csv('14 solar-share-energy.csv'),
    pd.read_csv('15 share-electricity-solar.csv'),
    pd.read_csv('16 biofuel-production.csv'),
    pd.read_csv('17 installed-geothermal-capacity.csv')
]

# Set the common columns (Entity, Code, and Year) as index for each dataframe
for i in range(len(dataframes)):
    dataframes[i].set_index(['Entity', 'Code', 'Year'], inplace=True)

# Concatenate the dataframes along columns (axis=1)
combined_datasets = pd.concat(dataframes, axis=1)

Combined_Data_complete = 'Combined_datasets.csv'
combined_datasets.reset_index(inplace=True)  # Reset index to have 'Entity', 'Code', and 'Year' as columns
combined_datasets.to_csv(Combined_Data_complete, index=False)
```

```python
import pandas as pd

# List of datasets
datasets = [
    '01 renewable-share-energy.csv',
    '02 modern-renewable-energy-consumption.csv',
    '03 modern-renewable-prod.csv',
    '04 share-electricity-renewables.csv',
    '05 hydropower-consumption.csv',
    '06 hydro-share-energy.csv',
    '07 share-electricity-hydro.csv',
    '08 wind-generation.csv',
    '09 cumulative-installed-wind-energy-capacity-gigawatts.csv',
    '10 wind-share-energy.csv',
    '11 share-electricity-wind.csv',
    '12 solar-energy-consumption.csv',
    '13 installed-solar-PV-capacity.csv',
    '14 solar-share-energy.csv',
    '15 share-electricity-solar.csv',
    '16 biofuel-production.csv',
    '17 installed-geothermal-capacity.csv'
]

# Load the first dataset to start with
final_df = pd.read_csv(datasets[0])

# Iterate over the rest of the datasets and merge them with the main dataframe
for dataset in datasets[1:]:
    temp_df = pd.read_csv(dataset)
    
    # Merging on common columns 'Entity', 'Code', and 'Year'
    final_df = pd.merge(final_df, temp_df, on=['Entity', 'Code', 'Year'], how='outer')

# Replace NaN values in the 'Code' column with the first three letters of the 'Entity' column in uppercase
final_df['Code'].fillna(final_df['Entity'].str[:3].str.upper(), inplace=True)

# Save the final dataframe
final_df.to_csv('combined_dataset.csv', index=False)

print("Completed merging and saving the dataset!")
```

```python
# Step 1: Set up the environment
# Make sure you've installed the necessary libraries before running the code
# pip install plotly dash pandas

# Step 2: Load the dataset and calculate averages
import pandas as pd

df = pd.read_csv('combined_dataset.csv')
average_consumption = df.groupby('Entity').mean().reset_index()

# Step 3: Create a Globe Visualization using Dash and Plotly
import dash
from dash import dcc, html
import plotly.express as px

# Initialize the Dash app
app = dash.Dash(__name__)

# Generate a choropleth map for renewable energy
fig = px.scatter_geo(average_consumption, 
                    locations="Entity",
                    locationmode='country names',
                    hover_name="Entity",
                    size="Renewables (% equivalent primary energy)", 
                    animation_frame="Year",
                    title="Average Consumption of Renewable Energies by Country",
                    template="plotly",
                    projection="orthographic")

# Dash layout for the app
app.layout = html.Div([
    dcc.Graph(figure=fig)
])

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
import pandas as pd

# Load the datasets
files = [
    '01 renewable-share-energy.csv',
    '02 modern-renewable-energy-consumption.csv',
    '03 modern-renewable-prod.csv',
    '04 share-electricity-renewables.csv',
    '05 hydropower-consumption.csv',
    '06 hydro-share-energy.csv',
    '07 share-electricity-hydro.csv',
    '08 wind-generation.csv',
    '09 cumulative-installed-wind-energy-capacity-gigawatts.csv',
    '10 wind-share-energy.csv',
    '11 share-electricity-wind.csv',
    '12 solar-energy-consumption.csv',
    '13 installed-solar-PV-capacity.csv',
    '14 solar-share-energy.csv',
    '15 share-electricity-solar.csv',
    '16 biofuel-production.csv',
    '17 installed-geothermal-capacity.csv',
    'COUNTRY_BUDGETS_1.csv',
    'COUNTRY_BUDGETS_2.csv',
    'COUNTRY_BUDGETS_3.csv'
]

datasets = [pd.read_csv(file) for file in files]

# Adjusting column names for the budget datasets
for i in range(17, 20):  # For ds18, ds19, and ds20
    datasets[i].rename(columns={'COUNTRY': 'Entity', 'TIME': 'Year'}, inplace=True)

# Merge all datasets based on 'Entity' and 'Year'
merged_df = datasets[0]

for ds in datasets[1:]:
    merged_df = merged_df.merge(ds, on=['Entity', 'Year'], how='outer')

# Handle NaN values in the Code column
merged_df['Code'] = merged_df['Code'].fillna(merged_df['Entity'].str[:3].str.upper())

# Calculate and add new columns
merged_df['Average Energy Production'] = merged_df[['Geo Biomass Other - TWh', 'Solar Generation - TWh', 'Wind Generation - TWh', 'Hydro Generation - TWh']].mean(axis=1)
merged_df['Average Energy Consumption'] = merged_df['Renewables (% equivalent primary energy)']

# Saving the combined dataset
merged_df.to_csv('combined_dataset.csv', index=False)
```
