```python
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("population.csv")
df = df.drop("Country Code", axis = 1)
df = df.drop("Indicator Code", axis = 1)
df = df.drop("Unnamed: 67", axis = 1)
df = df.drop("Indicator Name", axis = 1)
df["Info Category"] = "population"
df
```

```python
population_melted = pd.melt(
    df,
    id_vars=["Country Name", "Info Category"],
    var_name = "Year",
    value_name = "population"
)

population_melted
```

```python
population_melted.to_csv("Country_Population.csv")
```
